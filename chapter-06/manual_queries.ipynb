**SQLite** — не выдерживает больших нагрузок, на время одного обращения файл с БД блокируется.
Более того, встроенные методы не позволяют писать запросы с помощью ООП

```python
import sqlite3
```

```python
conn = sqlite3.connect("manual_queries.db")
cursor = conn.cursor()
```

```python
query = """
DROP TABLE IF EXISTS users;
"""
cursor.execute(query)

query = """
CREATE TABLE users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name VARCHAR(50),
    email VARCHAR(100)
);
"""
cursor.execute(query)
```

```python
query = """
INSERT INTO users
VALUES
    (NULL, "Анна Иванова", "anna@mail.ru"),
    (NULL, "Петр Сидоров", "petr@yandex.ru"),
    (NULL, "Мария Козлова", "maria@gmail.com");
"""
cursor.execute(query)
```

```python
query = """
SELECT * FROM users;
"""
cursor.execute(query)
users = cursor.fetchall()
print(users)
```

```python
query = """
SELECT * FROM users
WHERE name = "Анна Иванова"
LIMIT 1;
"""
cursor.execute(query)
users = cursor.fetchall()
print(users)
```

```python
query = """
UPDATE users
SET email = "anna_new@mail.ru"
WHERE name = "Анна Иванова";
"""
cursor.execute(query)
```

```python
query = """
DELETE FROM users
WHERE name = "Петр Сидоров";
"""
cursor.execute(query)
```

```python
query = """
SELECT * FROM users
"""
cursor.execute(query)
users = cursor.fetchall()
print(users)
```

```python
query = """
SELECT COUNT(*) FROM users
"""
cursor.execute(query)
users = cursor.fetchall()
print(users)
```

```python
query = """
SELECT *
FROM users
ORDER BY name;
"""
cursor.execute(query)
users = cursor.fetchall()
print(users)
```

```python
cursor.close()
conn.close()
```
